```python
import ollama
response = ollama.web_search("What is Ollama?")
print(response)
```

```python
for result in response.results:
        title = result.title
        url = result.url
        content = result.content
        
        # Print the data
        print(f"Title: {title}")
        print(f"URL: {url}")
        print(f"Content (snippet): {content[:200]}...") 
        print("-" * 30)
```

```python
# combining and summarizing all 

all_raw_content = []
    
print("Combining content from all search results...")
for result in response.results:
    # Get the raw content and add it to the list
    all_raw_content.append(result.content)


full_text_to_summarize = "\n".join(all_raw_content)

prompt_template = f"""
You are an expert summarizer. Your task is to read the provided text and give a concise summary of the main points.
**Crucially, you must ignore all links, URLs, and any navigation text** like "Sign In", "Open In App", "Explore", etc.

--- TEXT TO SUMMARIZE ---
{full_text_to_summarize}

--- CONCISE SUMMARY ---"""



# Call the generate function
summary_response = ollama.generate(
    model='llama3.2:1b',  # Or 'llama3', 'phi3'
    prompt=prompt_template
    # stream=False is the default
)

print("\n--- SUMMARY ---")
# The response text is in the 'response' key
print(summary_response['response'])
```
